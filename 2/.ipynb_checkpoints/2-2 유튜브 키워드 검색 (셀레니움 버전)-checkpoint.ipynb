```python
search_title = input('ê²€ìƒ‰ ì œëª©ì„ ìž…ë ¥í•´ì£¼ì„¸ìš”: ')
driver = webdriver.Chrome('/Users/dongyokim/main/jupyter/chromedriver/chromedriver')
driver.get('https://www.youtube.com/results?search_query={}'.format(search_title))
```

```python
# ì›í•˜ëŠ” ìœ„ì¹˜ ìŠ¤í¬ë¡¤ ë‚´ë¦¬ê¸°  
# finish_line = 40000 ê¸°ì¤€: 162 ê°œ
finish_line = 40000
last_page_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    # ìš°ì„  ìŠ¤í¬ë¡¤ ë‚´ë¦¬ê¸°
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(2.0)       # ìž‘ì—… ì¤‘ê°„ì— 1ì´ìƒìœ¼ë¡œ ê°„ê²©ì„ ì¤˜ì•¼ ë°ì´í„° ì·¨ë“ê°€ëŠ¥(ìŠ¤í¬ë¡¤ì„ ë‚´ë¦´ ë•Œì˜ ë°ì´í„° ë¡œë”© ì‹œê°„ ë•Œë¬¸)
    # í˜„ìž¬ ìœ„ì¹˜ ë‹´ê¸°
    new_page_height = driver.execute_script("return document.documentElement.scrollHeight")
    
    # ê³¼ê±°ì˜ ê¸¸ì´ì™€ í˜„ìž¬ ìœ„ì¹˜ ë¹„êµí•˜ê¸°
    if new_page_height > finish_line:
        break
    else: 
        last_page_height = new_page_height
```

```python
content_box = driver.find_elements_by_css_selector('#contents > ytd-video-renderer')
```

```python
youtube_list = []
for idx, box in enumerate(content_box):
    title = box.find_element_by_css_selector('#video-title > yt-formatted-string').text
    view_count = box.find_element_by_css_selector('#metadata-line > span:nth-child(1)').text.replace('ì¡°íšŒìˆ˜', "")
    past_day = box.find_element_by_css_selector('#metadata-line > span:nth-child(2)').text
    channel_name = box.find_elements_by_css_selector('#text > a')[1].text
    link = 'https://www.youtube.com/' + box.find_element_by_css_selector('#video-title').get_attribute('href')
    
    print("===== {}ë²ˆ box =====".format(idx+1))
    print('ì˜ìƒì œëª© :', title)
    print('ì¡°íšŒìˆ˜ :', view_count)
    print('ì˜ìƒ ê²Œì‹œì¼(í˜„ìž¬ë‚ ì§œê¸°ì¤€) :', past_day)
    print('ì±„ë„ëª… :', channel_name)
    print('ì˜ìƒ ë§í¬ :', link)
    print("====================", end='\n\n')
    
    
    youtube_list.append([title, view_count, past_day, channel_name, link])
    
    
    
    
    time.sleep(0.5)
print('{} í‚¤ì›Œë“œ ê²€ìƒ‰ ê²°ê³¼ ìˆ˜ì§‘ ì™„ë£Œ'.format(search_title))
```

```python
import pandas as pd 
youtube_list_df = pd.DataFrame(data=youtube_list, columns=['ì˜ìƒì œëª©', 'ì¡°íšŒìˆ˜', 'ì˜ìƒ ê²Œì‹œì¼(í˜„ìž¬ë‚ ì§œê¸°ì¤€)', 'ì±„ë„ëª…', 'ì˜ìƒ ë§í¬'])
```

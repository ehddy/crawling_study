```python
html = driver.page_source
soup = bs(html, 'html.parser')
```

```python
content_lists = soup.find('div', {'id':'contents'})
content_boxs = content_lists.find_all('ytd-video-renderer')
```

```python
youtube_list = []
for i, box in enumerate(content_boxs):
    title = box.select_one('#video-title > yt-formatted-string').text
    view_count = box.select_one('#metadata-line > span:nth-child(1)').text.replace('ì¡°íšŒìˆ˜', '')
    past_days = box.select_one('#metadata-line > span:nth-child(2)').text
    channel_name = box.select_one('#text > a').text
    link = 'https://www.youtube.com/' + box.select_one('#video-title')['href']
    
    
    print("===== {}ë²ˆ box =====".format(i+1))
    print('ì˜ìƒì œëª© :', title)
    print('ì¡°íšŒìˆ˜ :', view_count)
    print('ì˜ìƒ ê²Œì‹œì¼(í˜„ì¬ë‚ ì§œê¸°ì¤€) :', past_days)
    print('ì±„ë„ëª… :', channel_name)
    print('ì˜ìƒ ë§í¬ :', link)
    print('===================', end='\n\n')
    time.sleep(0.5)
    
    youtube_list.append([title, view_count, past_days, channel_name, link])

driver.quit()
```

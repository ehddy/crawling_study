## 게임소개 정보를 수집하여 자연어9장-11의 코드를 참고해 게임 추천 만드기

페이지를 넘어갈때 링크의 번화가 없다. 클릭을 통해 번거롭게 수집함.

각 페이지 링크 똑 같으면 어떻게 ????? 더 좋은 방법이 없을까??

```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
import requests
import pandas as pd
```

## 첫 페이지의 첫 번째 게임의 정보

```python
driver = webdriver.Chrome('chromedriver')
link = "https://www.gamemeca.com/game.php"
driver.get(link)

time.sleep(2)

#클릭할 버튼을 찾기
btn = driver.find_element(by = By.XPATH, value = '//*[@id="gmdb-list-box"]/ul/li[1]/a/span')
#클릭하여 상세피이지에 들어가기
btn.click()

for handle in driver.window_handles: #브라우저 탭 객체를 리스트로 변환
    driver.switch_to.window(handle) #탭으로 이동
    
    #현재 페이지의 링크
    url = driver.current_url
    html = requests.get(url)
    soup = bs(html.text, 'html.parser')
    html.close()
    #제목
    title = soup.select('div.content-left > div.gm_info_area > div > div.GmaeInfoTxt > h2')[0].text
    print(title)
    #이미지
    img = soup.select('div.content-left > div.gm_info_area > div > div.gameThumb > img')[0].get('src')
    print(img)
    #게임소개
    game_text = soup.select('div.content-left > div:nth-child(4) > div.db-cont1-left')[0].text
    print(game_text)
   
```

## 첫 페이지의 모든 게임의 정보

```python
driver = webdriver.Chrome('chromedriver')
link = "https://www.gamemeca.com/game.php"
driver.get(link)

url_list = []
title_list = []
img_list = []
text_list = []

#현재 페이지의 게임을 순환클릭하여 정보를 받아오기, 25개
btns= []
for num in range(1,26):
    # xpath 사용 할때 error. JavaScript 방법을 사용
    btn = 'document.querySelector("#gmdb-list-box > ul > li:nth-child(%d) > a > img").click()'%num
    btns.append(btn)

i = 0
while True:
    time.sleep(2)
    driver.execute_script(btns[i])
    
    #driver.switch_to.window(driver.current_window_handle)
    for handle in driver.window_handles:
        driver.switch_to.window(handle)
        #현재 페이지의 링크
        url = driver.current_url
        url_list.append(url)
        html = requests.get(url)
        soup = bs(html.text, 'html.parser')
        html.close()
        #제목
        title = soup.select('div.content-left > div.gm_info_area > div > div.GmaeInfoTxt > h2')[0].text
        print('제목은 "' + title +'"')
        title_list.append(title)
        #이미지
        img = soup.select('div.content-left > div.gm_info_area > div > div.gameThumb > img')[0].get('src')
        img_list.append(img)
        #게임소개
        game_text = soup.select('div.content-left > div:nth-child(4) > div.db-cont1-left')[0].text
        print("소개: " + game_text)
        text_list.append(game_text)
    print('---------------------------------------')
    time.sleep(3)
    driver.back()
    if i < 25:
        i += 1
    else:
        break
```

## page 1, 2, 3, 4....수집하기

```python
driver = webdriver.Chrome('chromedriver')
link = "https://www.gamemeca.com/game.php"
driver.get(link)

url_list = []
title_list = []
img_list = []
text_list = []

#현재 페이지의 게임을 순환클릭하여 정보를 받아오기, 25개
btns= []
for num in range(1,26):
    btn = 'document.querySelector("#gmdb-list-box > ul > li:nth-child(%d) > a > img").click()'%num
    btns.append(btn)

k = 0
while True:
    # k = 0 이면, page1
    if k == 0:
        i = 0
        while True:
            time.sleep(3)
            driver.execute_script(btns[i])
    
            #driver.switch_to.window(driver.current_window_handle)
            for handle in driver.window_handles:
                driver.switch_to.window(handle)
                time.sleep(1)
                #현재 페이지의 링크
                url = driver.current_url
                url_list.append(url)
                html = requests.get(url)
                soup = bs(html.text, 'html.parser')
                html.close()
                #제목
                title = soup.select('div.content-left > div.gm_info_area > div > div.GmaeInfoTxt > h2')[0].text
                title_list.append(title)
                #이미지
                img = soup.select('div.content-left > div.gm_info_area > div > div.gameThumb > img')[0].get('src')
                img_list.append(img)
                #게임소개
                game_text = soup.select('div.content-left > div:nth-child(4) > div.db-cont1-left')[0].text
                text_list.append(game_text)
    
            time.sleep(3)
            driver.back()
            if i < 24:
                i += 1
            else:
                break
    # page1뺴고 남머지 page들
    elif k > 0:
        click = k # k = 1 때, '다음'버튼을 click 1번하면 page2에 가기
        while click > 0:  
            next = 'document.querySelector("#list-page-nav > a.num_next.move-list-page").click()' 
            driver.execute_script(next)
            click -= 1
        #원하는 페이지에 들어가면, 크롤링 시작
            
        i = 0
        while True:
            time.sleep(3)
            driver.execute_script(btns[i])
    
            #driver.switch_to.window(driver.current_window_handle)
            for handle in driver.window_handles:
                driver.switch_to.window(handle)
                time.sleep(1)
                #현재 페이지 링크
                url = driver.current_url
                url_list.append(url)
                html = requests.get(url)
                soup = bs(html.text, 'html.parser')
                html.close()
                #제목
                title = soup.select('div.content-left > div.gm_info_area > div > div.GmaeInfoTxt > h2')[0].text
                title_list.append(title)
                #이미지
                img = soup.select('div.content-left > div.gm_info_area > div > div.gameThumb > img')[0].get('src')
                img_list.append(img)
                #게임소개 
                game_text = soup.select('div.content-left > div:nth-child(4) > div.db-cont1-left')[0].text
                text_list.append(game_text)
    
            time.sleep(3)
            # driver.back()하면 다 page1에 가기 때문에 '다음'번튼 클릭을 통해 다시 원하는 page에 가기
            #예를들어: page2의 첫 게임을 클롤링 끝나면, back하여, 다시 '다음'버튼을 클릭해야 함.
            driver.back()
            click = k # k = 2 때, '다음'버튼을 click 2번, page 3 에 가기 
            while click > 0:  
                next = 'document.querySelector("#list-page-nav > a.num_next.move-list-page").click()' 
                time.sleep(1)
                driver.execute_script(next)
                click -= 1
                
            if i < 24:
                i += 1
            else:
                break
        
    if k < 200: # page200까지 수집
        k += 1
        print('%d번째 페이지의 수집한 게임이름:'%k)
        print(title_list)
        time.sleep(5)
    else:
        break
```

```python
def get_page_one():
    i = 0
    while True:
        time.sleep(3)            
        driver.execute_script(btns[i])
    
        #driver.switch_to.window(driver.current_window_handle)
        for handle in driver.window_handles:
            driver.switch_to.window(handle)
            time.sleep(1)
            #현재 페이지의 링크
            url = driver.current_url
            url_info.append(url)
            html = requests.get(url)
            soup = bs(html.text, 'html.parser')
            html.close()
            #제목
            title = soup.select('div.content-left > div.gm_info_area > div > div.GmaeInfoTxt > h2')[0].text
            title_info.append(title)
            #이미지
            img = soup.select('div.content-left > div.gm_info_area > div > div.gameThumb > img')[0].get('src')
            img_info.append(img)
            #게임소개
            game_text = soup.select('div.content-left > div:nth-child(4) > div.db-cont1-left')[0].text
            text_info.append(game_text)
    
        time.sleep(3)
        driver.back()
        if i < 24:
            i += 1
        else:
            break
```

```python
def get_page():
    click = k # k = 1 때, '다음'버튼을 click 1번하면 page2에 가기
    while click > 0:  
        next = 'document.querySelector("#list-page-nav > a.num_next.move-list-page").click()' 
        driver.execute_script(next)
        click -= 1
        #원하는 페이지에 들어가면, 클롤링 시작
            
    i = 0
    while True:
        time.sleep(3)
        driver.execute_script(btns[i])
    
        #driver.switch_to.window(driver.current_window_handle)
        for handle in driver.window_handles:
            driver.switch_to.window(handle)
            time.sleep(1)
            #현재 페이지 링크
            url = driver.current_url
            url_info.append(url)
            html = requests.get(url)
            soup = bs(html.text, 'html.parser')
            html.close()
            #제목
            title = soup.select('div.content-left > div.gm_info_area > div > div.GmaeInfoTxt > h2')[0].text
            title_info.append(title)
            #이미지
            img = soup.select('div.content-left > div.gm_info_area > div > div.gameThumb > img')[0].get('src')
            img_info.append(img)
            #게임소개 
            game_text = soup.select('div.content-left > div:nth-child(4) > div.db-cont1-left')[0].text
            text_info.append(game_text)
    
        time.sleep(3)
        # driver.back()하면 다 page1에 가기 때문에 '다음'번튼 클릭을 통해 다시 원하는 page에 가기
        #예를들어: page2의 첫 게임을 크롤링 끝나면, back하여, 다시 '다음'버튼을 클릭해야 함.
        driver.back()
        click = k # k = 2 때, '다음'버튼을 click 2번, page 3 에 가기 
        while click > 0:  
            next = 'document.querySelector("#list-page-nav > a.num_next.move-list-page").click()' 
            time.sleep(1)
            driver.execute_script(next)
            click -= 1
                
        if i < 24:
            i += 1
        else:
            break
```

```python
driver = webdriver.Chrome('chromedriver')
link = "https://www.gamemeca.com/game.php"
driver.get(link)

url_info= []
title_info = []
img_info= []
text_info= []

#현재 페이지의 게임을 순환클릭하여 정보를 받아오기, 25개
btns= []
for num in range(1,26):
    btn = 'document.querySelector("#gmdb-list-box > ul > li:nth-child(%d) > a > img").click()'%num
    btns.append(btn)

k = 0
while True:
    # k = 0 이면, page1
    if k == 0:
        get_page_one()
    # page1뺴고 남머지 page들
    elif k > 0:
        get_page()
        
    if k < 200: # page200까지 수집
        k += 1
        print('%d번째 수집한 게임이름들:'%k)
        print(title_list)
        time.sleep(5)
    else:
        break
```

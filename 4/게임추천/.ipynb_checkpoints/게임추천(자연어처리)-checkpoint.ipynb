```python
stop_words = []
f = open('/content/drive/MyDrive/NLP/불용어.txt')
lines = f.readlines()
for line in lines:
  line = line.strip()
  stop_words.append(line)
```

```python
text = []
for i in  data['소개']:
  text.append(i)

text_df = pd.DataFrame({'text':text})
print(text_df[:5]) 
```

```python
# 특수 문자 제거
text_df['clean'] = text_df['text'].str.replace("[^ㄱ-ㅎ ㅏ-ㅣ 가-힣]", "")
print('특수 문자 제거:\n', text_df['clean'][:5])
```

```python
from konlpy.tag import Okt

okt = Okt()

# 토큰화
tokenized = text_df['clean'].apply(lambda x: okt.morphs(x))
print('토큰화 후:\n',tokenized[:5])
```

```python
clean= tokenized.apply(lambda x: [item for item in x if item not in stop_words])
print('불용어 제거:\n', clean[:5])
```

```python
corpus = []
for line in clean:
  corpus.append(line)
print(corpus[:5])
```

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
```

```python
loaded_model = KeyedVectors.load_word2vec_format("/content/drive/MyDrive/NLP/kor_w2v") # 모델 로드
```

```python
word2vec_model = loaded_model
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus) #빈도수 기준으로 단어 집합 생성
vocab_size = len(tokenizer.word_index) + 1 # 패딩을 고려하여 +1
print('단어 집합 :',vocab_size)
```

```python
x_encoded = tokenizer.texts_to_sequences(corpus) #정수인코딩
print('정수 인코딩 결과 :',x_encoded)

max_len = max(len(l) for l in x_encoded)
print('최대 길이 :',max_len)
```

```python
x_train = pad_sequences(x_encoded, maxlen=max_len, padding='post') #패딩
print(x_train)
```

```python
embedding_matrix = np.zeros((vocab_size, 100))
```

```python
def get_vector(word):
    if word in word2vec_model:
        return word2vec_model[word]
    else:
        return None
```

```python
for word, index in tokenizer.word_index.items():
    # 단어와 맵핑되는 사전 훈련된 임베딩 벡터값
    vector_value = get_vector(word)
    if vector_value is not None:
        embedding_matrix[index] = vector_value
```

```python
print(word2vec_model['히트'])
```

```python
def get_document_vectors(document_list):
    document_embedding_list = []

    # 각 문서에 대해서
    for line in document_list:
        doc2vec = None
        count = 0
        for word in line: #각 문서의 단어
            if word in word2vec_model.wv.vocab:
                count += 1  
                # 해당 문서에 있는 모든 단어들의 벡터값을 더한다.
                if doc2vec is None:
                    doc2vec = word2vec_model[word] #단어의 벡터값
                else:
                    doc2vec = doc2vec + word2vec_model[word] #각 문서에 존재하는 단어들의 벡터값

        if doc2vec is not None:
            # 단어 벡터를 모두 더한 벡터의 값을 문서 길이로 나눠준다.
            doc2vec = doc2vec / count #평균
            document_embedding_list.append(doc2vec)

    # 각 문서에 대한 문서 벡터 리스트를 리턴
    return document_embedding_list
```

```python
clean_df = pd.DataFrame({'clean_df':clean})
clean_df
```

```python
document_embedding_list = get_document_vectors(clean_df['clean_df'])
```

```python
print('문서 벡터의 수 :',len(document_embedding_list))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarities = cosine_similarity(document_embedding_list, document_embedding_list)
```

```python
cosine_similarities[0]
```

```python
print('코사인 유사도 매트릭스의 크기 :',cosine_similarities.shape)
```

```python
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO
```

```python
from pandas.core.base import DataError
def recommendations(title):
    games = data[['게임이름', '이미지']]

    # 게임의 제목을 입력하면 해당 제목의 인덱스를 리턴받아 idx에 저장.
    indices = pd.Series(data.index, index = data['게임이름']).drop_duplicates() 
    idx = indices[title]

    # 입력된 책과 줄거리(document embedding)가 유사한 책 5개 선정.
    sim_scores = list(enumerate(cosine_similarities[idx]))
    sim_scores = sorted(sim_scores, key = lambda x: x[1], reverse = True)
    sim_scores = sim_scores[1:6]

    # 가장 유사한 책 5권의 인덱스
    game_indices = [i[0] for i in sim_scores]

    # 전체 데이터프레임에서 해당 인덱스의 행만 추출. 5개의 행을 가진다.
    recommend = games.iloc[game_indices].reset_index(drop=True) 

    fig = plt.figure(figsize=(20, 30))

    # 데이터프레임으로부터 순차적으로 이미지를 출력
    for index, row in recommend.iterrows():
        response = requests.get(row['이미지'])
        img = Image.open(BytesIO(response.content))
        fig.add_subplot(1, 5, index + 1)
        plt.imshow(img)
        plt.title(row['게임이름'])
```

```python
recommendations("더 루커")
```

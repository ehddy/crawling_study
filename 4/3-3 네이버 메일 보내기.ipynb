```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication 
```

```python

def naver_mail_send():
    
    smtpName = "smtp.naver.com" #smtp 서버 주소
    smtpPort = 587 #smtp 포트 번호

    sendEmail = input('사용하실 이메일 주소를 입력해주세요 : ')
    print()
    password = input('비밀번호를 입력해주세요 : ')
    print()
    recvEmail = input('상대방의 이메일 주소를 입력해주세요 : ')
    print()
    
    #여러 MIME을 넣기위한 MIMEMultipart 객체 생성
    msg = MIMEMultipart()
    
    print('==========================================')
    
    title = input('메일 제목을 입력해주세요 : ')
    print()
    text = input('메일 내용을 입력해주세요 : ')
    print()
        
#     msg = MIMEText(text) #MIMEText(text , _charset = "utf8")

    msg['Subject'] = title 
    msg['From'] = sendEmail
    msg['To'] = recvEmail
    
    #본문 추가
    contentPart = MIMEText(text) #MIMEText(text , _charset = "utf8")
    msg.attach(contentPart) 

    
    #파일 추가
    file_y_or_n = input('파일을 첨부하시겠습니까?(yes or no) : ')
    
    if file_y_or_n == 'yes':
        etcFileName = input('파일경로나 파일명을 입력해주세요. : ')
        with open(etcFileName, 'rb') as etcFD : 
            etcPart = MIMEApplication( etcFD.read() )
            #첨부파일의 정보를 헤더로 추가
            etcPart.add_header('Content-Disposition','attachment', filename=etcFileName)
            msg.attach(etcPart) 
            print('파일을 첨부했습니다. ')
            print()


    
    print(msg.as_string())
    print()
    
    s=smtplib.SMTP( smtpName , smtpPort ) #메일 서버 연결
    print('메일 서버를 연결했습니다.')
    print()
    
    s.starttls() #TLS 보안 처리
    print('보안 처리를 완료했습니다.')
    print()
    
    
    s.login( sendEmail , password ) #로그인
    print('사용자의 계정에 로그인했습니다.')
    print()
    print('========================')
    
    s.sendmail( sendEmail, recvEmail, msg.as_string() ) #메일 전송, 문자열로 변환하여 보냅니다.
    print('메일을 보냈습니다. ')
    print()
    
    s.close() #smtp 서버 연결을 종료합니다.
    print('========================')
    print('SMTP 서버 연결을 종료합니다.')
    
    return 
```

## 1. 구글 메일 보내기 

---

### 구글 메일 서버
메일 서버를 이용하려면 서버명과 포트 및 보안방식을 알아야 한다 



- 서버명 : smtp.gmail.com
- 인증 필요: 예
- TLS/STARTTLS용 포트: 587

### 메일 보내기 
---
SMTP(Simple Mail Transfer Protocol) : 단순히 영어 내용만 있는 메일을 전송할 때  
MIME(Multipurpose Internet Mail Extensions) : 다국어 지원, 파일 첨부 등의 기능이 현재는 당연하게 사용  

```python
import smtplib
from email.mime.text import MIMEText
```

```python
sendEmail = "********8"
recvEmail = "********"
password = "*********"

smtpName = "smtp.gmail.com" #smtp 서버 주소
smtpPort = 587 #smtp 포트 번호

text = '밥먹자 동요야'
msg = MIMEText(text) #MIMEText(text , _charset = "utf8")

msg['Subject'] ="안녕하신가요"
msg['From'] = sendEmail
msg['To'] = recvEmail
print(msg.as_string())

s=smtplib.SMTP( smtpName , smtpPort ) #메일 서버 연결
s.starttls() #TLS 보안 처리
s.login( sendEmail , password ) #로그인
s.sendmail( sendEmail, recvEmail, msg.as_string() ) #메일 전송, 문자열로 변환해야 합니다.
s.close() #smtp 서버 연결을 종료합니다.
```

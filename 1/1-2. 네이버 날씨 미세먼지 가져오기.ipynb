```python
from bs4 import BeautifulSoup as bs 
from pprint import pprint
import requests

from selenium import webdriver
import time

driver = webdriver.Chrome("/Users/dongyokim/main/jupyter/chromedriver/chromedriver")
driver.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EB%82%A0%EC%94%A8")
driver.implicitly_wait(3)    
driver.set_window_size(1200, 800)
```

```python
# 위에 페이지에서 미세먼지와 초미세먼지에 대한 상세적인 수치가 나와있지 않아, 셀레니움을 이용하여 현재 위치의 다양한 날씨 정보를 나타내는 사이트로 이동해준다.
link = driver.find_element_by_xpath('//*[@id="main_pack"]/section[1]/div[1]/div[2]/div[1]/div[1]/div/div[2]/div/div[2]/ul/li[1]/a').get_attribute('href')
driver.find_element_by_xpath('//*[@id="main_pack"]/section[1]/div[1]/div[2]/div[1]/div[1]/div/div[2]/div/div[2]/ul/li[1]/a').click()
```

```python
# 상세 날씨 정보를 포함한 사이트의 링크 수집 
res = requests.get(link)

# 파이썬에서 쉽게 분석할 수 있도록 파싱 작업 
soup = bs(res.text, 'html.parser')
```

### HTML 요소는 <태그 속성 = 속성값> 텍스트가 기본 구조

- 태그 : div
- 속성 : class
- 속성값 : detail_content

```python
# 현재 위치 텍스트 추출 
log_data = soup.find('div', {'class':'title_area'}).find('h2', {'class':'blind'}).text

# 현재 날짜 및 시간 텍스트 추출 
time_data = soup.find('span', {'class':'time'}).find('em').text
real_time = time_data.split(' ')


# 미세먼지, 초미세먼지 리스트(수치, 수준) 도출 
box_data = soup.find('div',{'class':'detail_content'}).find_all('div', {'class':'grade'})
```

```python
# 미세먼지와 초미세먼지 값 도출 
count_data = [int(box_data[i].find('span', {'class':'num'}).text) for i in range(len(box_data))]
# 미세먼지 및 초미세먼지 값 수준 
status_data = [box_data[i].find('span', {'class':'text'}).text for i in range(len(box_data))]

print(count_data)
print(status_data)
```

```python
print("조회 날짜: {}\n조회 시간: {}".format(real_time[0], real_time[1]))
print('현재 위치하신 {}의 미세먼지 농도 수치는 {}로 "{}" 수준이며, 초미세먼지 수치는 {}로 "{}" 수준입니다.'.format(log_data, count_data[0], status_data[0], count_data[1], status_data[1]))
```

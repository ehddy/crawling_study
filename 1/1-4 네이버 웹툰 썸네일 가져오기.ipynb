```python
from bs4 import BeautifulSoup as bs 
import requests
import pandas as pd

from urllib.request import urlretrieve

link = "https://comic.naver.com/webtoon/weekday.nhn"

res = requests.get(link)

soup = bs(res.text, 'html.parser')
```

```python
col_list = soup.select('div.list_area.daily_all > div.col')
```

```python
col_list[2].select('div.col_inner > ul > li > a')
```

```python
weekdays = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
```

```python
total_name_lists = []
total_img_lists = []

for idx, col in enumerate(col_list):
    weekday_list = col.select('div.col_inner > ul > li')
    total_name_list = []
    total_img_list = []
    print("===== {} 웹툰 리스트 =====".format(weekdays[idx]))
    for day in weekday_list:
        name = day.select_one('a:nth-child(2)').text 
        img_link = day.select_one('a:nth-child(1) > img')['src']
        print(name)
        total_name_list.append(name)
        total_img_list.append(img_link)
    total_name_lists.append(total_name_list)
    total_img_lists.append(total_img_list)
    print()
    print('============')
    print()
```

```python
# 행렬 바꾸기 
new_list = list(map(list, zip(*total_name_lists)))
```

```python
df = pd.DataFrame(data=new_list, columns=weekdays)
```

```python
df.to_excel('웹툰 제목 리스트.xlsx')
```
